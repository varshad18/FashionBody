```python
def getClass():
  clnamesplit=clname.split("_")
  regVal=clnamesplit[-1]
  #regVal=clname[-1]
  regVal=float(regVal)
  #regVal
  #clname
  #clnamesplit [0]Shoulder_[1]Breadth_[2]4.0
  clas=clnamesplit[0]+'_'+clnamesplit[1]
  result=[]
  result.append(clas)
  result.append(regVal)
  #print("inside function")
  #print("result=", result)
  return result
```

```python
#to obtain filenames with no repetition
def uniquefn(filenames):
  for i in filenames:
    if i not in uniquefilenames:
      uniquefilenames.append(i)
  return uniquefilenames
#filenames
#uniquefilenames
```

```python
#To parse filename and make it shorter
def shortfilename(filename):
  filen=filename.split("_",1)
  ext=filen[1].split(".")
  filen1=filen[0]
  ext1=ext[-1]
  #print("filen=",filen1)
  #print("ext=",ext1)
  newfilename=filen1+'.'+ext1

  return newfilename
```

```python
def create_img(imgpath, reg_part, x1, y1, x2, y2):
  img = Image.open(imgpath)
  #display.display(img)
  parts=imgpath.split(".")
  try:
    os.mkdir("/content/new")
  except OSError as error:
    print(error)
  savpath="/content/new/"+parts[0]+"_"+reg_part+"."+parts[1]
  #print(parts)
  print(savpath)
  box = (x1,y1,x2,y2)
  #print(box)
  img2 = img.crop(box)
  #display.display(img2)
  img2.save(savpath)
  #img2.show()
```
