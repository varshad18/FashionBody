```python
import xml.etree.ElementTree as ET
import os

# Specify the directory path (replace with your actual path)
directory_path = "/content/drive/MyDrive/FashionBody/FashionBodyPartDetection/OriginalData_1Arm/NewnowAnnotations/"

# Required names and their expected counts
required_names = {
  "Arm": 2,
  "Chest": 1,
  "Hips": 1,
  "Legs": 1,
  "LegsBred": 1,
  "Shoulders": 1,
  "UpArm": 2,
  "Waist": 1,
  "whole": 1
}

# List to store filenames with missing or incorrect name tag counts
invalid_files = []


for filename in os.listdir(directory_path):
  # Check if it's a valid XML file
  if os.path.isfile(os.path.join(directory_path, filename)) and filename.endswith(".xml"):
    try:
      # Parse the XML file
      tree = ET.parse(os.path.join(directory_path, filename))
      root = tree.getroot()

      # Count occurrences of each name tag
      name_counts = {}
      for element in root.iter():
        if element.tag == "name":
          name = element.text
          name_counts[name] = name_counts.get(name, 0) + 1

      # Check for missing names or invalid counts
      missing_names = [name for name in required_names if name not in name_counts]
      invalid_counts = [(name, count) for name, count in name_counts.items() if count != required_names[name]]

      if missing_names or invalid_counts:
        invalid_files.append(filename)
        print(f"File: {filename}")
        if missing_names:
          print(f"- Missing names: {', '.join(missing_names)}")
        if invalid_counts:
          for name, count in invalid_counts:
            expected_count = required_names[name]
            print(f"- Name '{name}' has {count} occurrences (expected: {expected_count})")

    except FileNotFoundError:
      print(f"Error: File '{filename}' not found. Skipping.")

if not invalid_files:
  print("All XML files have the expected names and counts in the 'name' tag.")
else:
  print(f"\n{len(invalid_files)} files have invalid name tags:")
```

```python
import os

def has_matching_annotation(image_name, annotations_dir):
  """
  This function checks if an image has a corresponding annotation file.

  Args:
      image_name: The name of the image file (including extension).
      annotations_dir: The path to the directory containing annotation files.

  Returns:
      True if a matching annotation file is found, False otherwise.
  """
  # Extract the filename without extension
  image_filename = os.path.splitext(image_name)[0]

  # Construct the expected annotation filename (assuming same name with .xml extension)
  annotation_filename = f"{image_filename}.xml"  # You can modify the extension if needed

  # Check if the annotation file exists
  annotation_path = os.path.join(annotations_dir, annotation_filename)
  return os.path.exists(annotation_path)

def check_image_annotations(img_dir, annotations_dir):
  """
  This function checks if all images have matching annotations.

  Args:
      img_dir: The path to the directory containing images.
      annotations_dir: The path to the directory containing annotation files.
  """
  for filename in os.listdir(img_dir):
    # Check if it's an image file
    if not filename.lower().endswith((".jpg", ".jpeg", ".png")):
      continue

    if not has_matching_annotation(filename, annotations_dir):
      print(f"Image '{filename}' does not have a matching annotation file.")

# Example usage
img_directory = "/content/drive/MyDrive/FashionBody/FashionBodyPartDetection/OriginalData_1Arm/updatedimg/"
annotations_directory = "/content/drive/MyDrive/FashionBody/FashionBodyPartDetection/OriginalData_1Arm/NewnowAnnotations/"

check_image_annotations(img_directory, annotations_directory)

print("Finished checking image annotations.")
```
